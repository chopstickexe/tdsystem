```python
import matplotlib.pyplot as plt
import pickle
import sys

sys.path.append("../") # go to parent dir

from tdsystem.record_page_parser import Record, RecordPageParser

with open('../records.pickle', 'rb') as f:
    rs = pickle.load(f)

records = []
for r in rs:
    records.append(r.record.total_seconds())

# the histogram of the data
n, bins, patches = plt.hist(records)
plt.xticks([150, 200, 250, 300, 350, 400], ['2:30', '3:20', '4:10', '5:00', '5:50', '6:40']) 
plt.ylabel('Number of records')
plt.title('Women 200IM (All Ages)')

plt.show()
```

```python
fly = []
ba = []
br = []
fr = []
for r in rs:
    avg = r.record.total_seconds()/4
    fly.append(r.lap[0].total_seconds() - avg)
    ba.append(r.lap[1].total_seconds() - r.lap[0].total_seconds()  - avg)
    br.append(r.lap[2].total_seconds() - r.lap[1].total_seconds() - avg)
    fr.append(r.lap[3].total_seconds() - r.lap[2].total_seconds() - avg)
    
# the histogram of the data
plt.hist(fly)
plt.yticks(range(0, 30, 5)) 
plt.xticks(range(-30, 31, 5)) 
plt.ylabel('Number of records')
plt.title('Women 200IM Fly lap - 50m avg (All Ages)')
plt.show()

# the histogram of the data
plt.hist(ba)
plt.yticks(range(0, 30, 5)) 
plt.xticks(range(-30, 31, 5)) 
plt.ylabel('Number of records')
plt.title('Women 200IM Ba lap - 50m avg (All Ages)')
plt.show()

# the histogram of the data
plt.hist(br)
plt.yticks(range(0, 30, 5)) 
plt.xticks(range(-30, 31, 5)) 
plt.ylabel('Number of records')
plt.title('Women 200IM Br lap - 50m avg (All Ages)')
plt.show()

# the histogram of the data
plt.hist(fr)
plt.yticks(range(0, 30, 5)) 
plt.xticks(range(-30, 31, 5)) 
plt.ylabel('Number of records')
plt.title('Women 200IM Fr lap - 50m avg (All Ages)')
plt.show()
```
